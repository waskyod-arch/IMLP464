# 資料前處理(Label encoding、 One hot encoding)
這兩個編碼方式的目的是為了將類別 (categorical)或是文字(text)的資料轉換成數字，而讓程式能夠更好的去理解及運算。
> Label encoding : 把每個類別 mapping 到某個整數，不會增加新欄位

> One hot encoding : 為每個類別新增一個欄位，用 0/1 表示是否

![](images/Encoder.PNG)


## Encoding Categorical features (or label)
![](images/Encoding.PNG)


````python
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame({'blood':['A','B','AB','O','B'], 
                   'Y':['high','low','high','mid','mid'],
                   'Z':[np.nan,np.nan,-1196,72,83]});
df
````

# 方法一：sklearn - label encoder + onehot encoder
>onehot encoder要用2D array，若維度所以要用reshape(-1,1)<br>
>onehot encoder要數字，若資料文文字要先用label encoder轉數字

````python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoder_Y=encoder.fit_transform(df['blood'])
print(encoder_Y)
df["blood"]=encoder_Y #因為轉換後有順序，可以直接丟回去frame!
df

"""您可能會想：「為什麼不能直接用 OneHotEncoder().fit(...) 就好，非得要取個名字叫 onehot 呢？」

這牽涉到機器學習裡非常重要的一個觀念：狀態與記憶 (State & Memory)。

這台 onehot 機器在運作時分為兩個階段：

學習階段 (Fit)： 當您給它看您的資料（例如：台北、台中、高雄），這台 onehot 機器會在它的肚子裡**「記住」**這些類別的順序（例如：台北是第 0 欄，台中是第 1 欄）。

轉換階段 (Transform)： 接下來，您要它把資料轉換成 0 和 1 的矩陣。

如果您沒有用一個變數（像是 onehot）把它存起來，這台機器在做完第一步（看資料）之後，就會立刻被電腦的資源回收車載走銷毀！等您要進行第二步轉換時，它已經忘記剛才學過的對照表了。

👉 結論： 寫下 onehot = OneHotEncoder()，就是為了把這台具有「學習記憶」的實體機器留在您的桌面上，隨時聽候您的差遣！"""
````

````python
from sklearn.preprocessing import OneHotEncoder
onehot=OneHotEncoder() #Onehotendoing只能用再array 所以要轉檔
print(type(df["blood"]))
d=np.array(df["blood"])
d.shape
onehot_df=onehot.fit_transform(d.reshape(-1,1)).toarray() #OneHotEncoder會轉出scipy.csr_matrix資料結構用.toarray()轉array
onehot_df
````

OneHotEncoder會轉出scipy.csr_matrix資料結構用.toarray()轉array
從結果可以知道，數字0的column 代表的是A、數字1的column 代表的是B，而數字2的column 代表的是AB。
除了轉換字串外，One hot encoding也可以轉換數字。在此處的data就不需要先經過Label encoding編碼

# 方法二：Keras - label encoder + to_categorical
>to_categorical要數字，若資料文文字要先用label encoder轉數字

````python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical # 修正 Keras 匯入語法

df = pd.DataFrame({'blood':['A','B','AB','O','B'], 
                   'Y':['high','low','high','mid','mid'],
                   'Z':[np.nan,np.nan,-1196,72,83]});

# label encoder 
encoder=LabelEncoder()
encoder_Y=encoder.fit_transform(df['blood'])
print(encoder_Y)
df["blood"]=encoder_Y #因為轉換後有順序，可以直接丟回去frame!
df

# convert integers to one hot encoding
"""keras_onehot=np_utils.to_categorical(encoder_Y)
keras_onehot"""
keras_onehot = to_categorical(encoder_Y)
print("\nOne-Hot Encoding 轉換結果:")
print(keras_onehot)
````

## 練習一：sklearn - label encoder + onehot encoder
下面的資料可以看到country那欄皆為字串， 大部分的模型都是基於數學運算，字串無法套入數學模型進行運算，<br>
在此先對其進行Label encoding編碼，我們從 sklearn library中導入 LabelEncoder class，對第一行資料進行fit及transform並取代之。

````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

country=['Taiwan','Australia','Ireland','Australia','Ireland','Taiwan']
age=[25,30,45,35,22,36]
salary=[20000,32000,59000,60000,43000,52000]
dic={'Country':country,'Age':age,'Salary':salary}
data=pd.DataFrame(dic)
data


encoder=LabelEncoder()
encoder_Y=encoder.fit_transform(data['Country'])
print(encoder_Y)
data["Country"]=encoder_Y #因為轉換後有順序，可以直接丟回去frame!

onehot=OneHotEncoder() #Onehotendoing只能用再array 所以要轉檔
print(type(data["Country"]))
d=np.array(data["Country"])
d.shape
onehot_df=onehot.fit_transform(d.reshape(-1,1)).toarray() #OneHotEncoder會轉出scipy.csr_matrix資料結構用.toarray()轉array
onehot_df

columnTransformer = ColumnTransformer([('encoder', 
                                        OneHotEncoder(), 
                                        [0])], 
                                      remainder='passthrough')  #有沒有reminder會刪除其他沒有執行的變項
data = np.array(columnTransformer.fit_transform(df), dtype = str) 
print(data)

data_le=pd.DataFrame(data)
print(data_le)
````

## 練習二：Keras - label encoder + to_categorical

````python
from sklearn.preprocessing import LabelEncoder
#from keras.utils import np_utils
from tensorflow.keras.utils import to_categorical # 修正 Keras 匯入語法

country=['Taiwan','Australia','Ireland','Australia','Ireland','Taiwan']
age=[25,30,45,35,22,36]
salary=[20000,32000,59000,60000,43000,52000]
dic={'Country':country,'Age':age,'Salary':salary}
data=pd.DataFrame(dic)
print(data)


# label encoder 
encoder=LabelEncoder()
encoder_Y=encoder.fit_transform(data['Country'])
print(encoder_Y)
data["Country"]=encoder_Y #因為轉換後有順序，可以直接丟回去frame!
print(data)

# convert integers to one hot encoding
"""keras_onehot=np_utils.to_categorical(encoder_Y)
keras_onehot"""
keras_onehot = to_categorical(encoder_Y)
print("\nOne-Hot Encoding 轉換結果:")
print(keras_onehot)

# --- 以下是放回 DataFrame 的新增步驟 ---
# 1. 取得原來的類別名稱 (這會自動依照 0, 1, 2 對應 'Australia', 'Ireland', 'Taiwan')
category_names = encoder.classes_

# 2. 將 keras_onehot 陣列轉換成新的 DataFrame，並為欄位命名 (例如: Country_Taiwan)
onehot_df = pd.DataFrame(keras_onehot, columns=[f"Country_{name}" for name in category_names])

# 3. 刪除原本會造成模型誤判的 'Country' 數字欄位
data = data.drop('Country', axis=1)

# 4. 將原本的 data 與新的 onehot_df 橫向合併 (axis=1 代表左右合併)
final_data = pd.concat([data, onehot_df], axis=1)

print("\n最終合併後的 DataFrame:")
print(final_data)

````
