## 1. 陣列導向 101

科學計算一個很核心的概念叫 "array oriented" 的寫法。Array 是 `numpy` 標準的資料結構, 和 list 很像, 但就差了那麼一點點。而這一點點讓我們在計算上是無比的方便。

### 【暖身】 計算平均

某位同學期中考各科成績如下, 請幫他計算成績。

    grades = [77, 85, 56, 90, 66]
    
請計算平均。

```python
s=0
grades = [77, 85, 56, 90, 66]
for i in grades:
    s+=i
print ("average = {0:6.2f}, sum = {1}".format(s/len(grades),s))
#print(f"average = {s/len(grades):6.2f}, sum = {s}")
"""簡單來說，在 .format() 的世界裡，大括號 {} 就像是**「叫號機」，而它後面括號裡的東西就是在「排隊的客人」**。

我們來拆解您的這行程式碼：
print("average = {0:6.2f}, sum = {1}".format(s/len(grades), s))

1. {0} 的 0 是什麼意思？
在程式設計的世界裡，我們數數字通常是從 0 開始數的。
所以 .format() 後面的括號裡，排隊的順序是這樣的：

0 號客人： s/len(grades) (這是平均值)

1 號客人： s (這是總和)

當字串裡寫著 {0} 時，Python 就會去叫 0 號客人 優先入座。也就是把 s/len(grades) 算出來的結果填進去。（後面的 :6.2f 則是順便規定這位客人要穿什麼衣服：總共佔 6 個字元寬，並且保留 2 位小數）。

2. sum 為什麼接 {1}，而不是直接寫 {s}？
因為這整句話是用雙引號 " " 包起來的字串。在雙引號裡面，Python 只會把它當作「純文字」，它根本不認識 s 是個變數。

如果您直接寫 {s}，Python 會去 .format() 裡面找有沒有一位名字叫 s 的客人，結果找不到，就會報錯。

所以我們必須用**「位置號碼」**來叫人： {1} 的意思就是「請把排在第 1 號位置的客人（也就是變數 s）請到這個位子來」。"""
```

## 5. 超重要 `axis` 觀念

初學 `numpy` 很多人有點弄不清楚 `axis` 概念。其實掌握矩陣, 或很像矩陣的陣列都是「先列(row)後行(col)」就可以!

我們先弄個 array 來練習。

```python
A=np.arange(10).reshape(2,5)
A
```

### 【重點】 一列一列算下來是 `axis=0`

![axis=0](images/axis0.png)

```python
a_sum=A.sum(axis=0)
print(a_sum)
```

### 【重點】 一行一行算過去是 `axis=1`

![axis=1](images/axis1.png)

```python
a_sum=A.sum(axis=1)
print(a_sum)
```

### 【提示】當然也有可能全部算

```python
A.sum().item()
```

## 7. 次元切割刀

`numpy` 中 array 的切割法和 list 很像。

```python
x=np.arange(10)
x
```

```python
x[3:7]
```

### 【技巧】2維陣列切法

記得先列後行!

```python
x=x.reshape(2,5)
```

要所有的row, 切出行 1-3 位置。

```python
x1=x[:, 1:3]
print(x1)
```

要所有的行col, 切出第 1 列row!

```python
x2=x[1,:]
print(x2)
```
